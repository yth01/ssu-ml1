```python
# 외부 데이터 다운로드
sp500 = yf.download('^GSPC', start='2018-12-01', end='2024-01-01')
usdkrw = yf.download('KRW=X', start='2018-12-01', end='2024-01-01')

#  MultiIndex 컬럼 평탄화
sp500.columns = sp500.columns.droplevel(1)  # ('Close', '^GSPC') → 'Close'
usdkrw.columns = usdkrw.columns.droplevel(1)

# 필요한 컬럼만 선택 및 리네임
sp500 = sp500[['Close']].rename(columns={'Close': 'SP500_Close'}).reset_index()
usdkrw = usdkrw[['Close']].rename(columns={'Close': 'USD_KRW'}).reset_index()

# 날짜 타입 통일 (normalize로 시간 제거)
df_train['Date'] = pd.to_datetime(df_train['Date']).dt.normalize()
df_test['Date'] = pd.to_datetime(df_test['Date']).dt.normalize()
sp500['Date'] = pd.to_datetime(sp500['Date']).dt.normalize()
usdkrw['Date'] = pd.to_datetime(usdkrw['Date']).dt.normalize()

# Merge
df_train = pd.merge(df_train, sp500, on='Date', how='left')
df_train = pd.merge(df_train, usdkrw, on='Date', how='left')

df_test = pd.merge(df_test, sp500, on='Date', how='left')
df_test = pd.merge(df_test, usdkrw, on='Date', how='left')

# 결측치 처리
df_train.fillna(method='ffill', inplace=True)
df_test.fillna(method='ffill', inplace=True)
df_train.fillna(method='bfill', inplace=True)
df_test.fillna(method='bfill', inplace=True)
```

```python
df_train['target'] = df_train['Close'].shift(-1)
df_test['target'] = df_test['Close'].shift(-1)
df_train = df_train[:-1].reset_index(drop=True)
df_test = df_test[:-1].reset_index(drop=True)

def create_windowed_data(df, n_days):
    feature_cols = ['Open', 'High', 'Low', 'Close', 'Volume', 'SP500_Close', 'USD_KRW']
    X, y = [], []
    for i in range(len(df) - n_days):
        window = df[feature_cols].iloc[i:i + n_days].values.flatten()
        target = df['target'].iloc[i + n_days]
        X.append(window)
        y.append(target)
    return np.array(X), np.array(y)

n_days = 3  # 이전과 동일하게 3일

X_train_scaled, y_train_scaled = create_windowed_data(df_train, n_days)
X_test_scaled, y_test_scaled = create_windowed_data(df_test, n_days)
dates_test_scaled = df_test['Date'].values[n_days:]

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_scaled)
X_test_scaled = scaler.transform(X_test_scaled)

X_train_raw = df_train[['Open', 'High', 'Low', 'Close', 'Volume', 'SP500_Close', 'USD_KRW']].values
y_train_raw = df_train['target'].values
X_test_raw = df_test[['Open', 'High', 'Low', 'Close', 'Volume', 'SP500_Close', 'USD_KRW']].values
y_test_raw = df_test['target'].values
dates_test_raw = df_test['Date'].values

models = {
    'LinearRegression': LinearRegression(),
    'Ridge': Ridge(alpha=1.0),
    'Lasso': Lasso(alpha=0.1),
    'RandomForest': RandomForestRegressor(n_estimators=100, random_state=42),
    'XGBoost': XGBRegressor(n_estimators=100, random_state=42, verbosity=0)
}

kf = KFold(n_splits=5, shuffle=True, random_state=42)

def evaluate_model(X_train, y_train, X_test, y_test, dates_test, preprocessing_label):
    cv_results = {}
    test_results = {}
    test_predictions = {}

    for name, model in models.items():
        fold_mae, fold_rmse = [], []

        for train_idx, val_idx in kf.split(X_train):
            X_tr, X_val = X_train[train_idx], X_train[val_idx]
            y_tr, y_val = y_train[train_idx], y_train[val_idx]

            model.fit(X_tr, y_tr)
            y_val_pred = model.predict(X_val)
            fold_mae.append(mean_absolute_error(y_val, y_val_pred))
            fold_rmse.append(np.sqrt(mean_squared_error(y_val, y_val_pred)))

        avg_mae = np.mean(fold_mae)
        avg_rmse = np.mean(fold_rmse)
        cv_results[name] = {'CV_MAE': avg_mae, 'CV_RMSE': avg_rmse}

        model.fit(X_train, y_train)
        y_test_pred = model.predict(X_test)
        test_results[name] = {'Test_MAE': mean_absolute_error(y_test, y_test_pred),
                              'Test_RMSE': np.sqrt(mean_squared_error(y_test, y_test_pred))}
        test_predictions[name] = y_test_pred

        print(f"\n[{preprocessing_label}] {name} 결과")
        print(f"Cross Validation - MAE: {avg_mae:.4f}, RMSE: {avg_rmse:.4f}")
        print(f"Test Set          - MAE: {test_results[name]['Test_MAE']:.4f}, RMSE: {test_results[name]['Test_RMSE']:.4f}")

    return test_predictions, test_results

print("\n======== Part 2: 전처리 적용 (Scaling + Windowing with External Features) ========")
test_predictions_scaled, test_results_scaled = evaluate_model(X_train_scaled, y_train_scaled, X_test_scaled, y_test_scaled, dates_test_scaled, "Scaled + Windowing (External)")

print("\n======== Part 2: 전처리 미적용 (No Preprocessing with External Features) ========")
test_predictions_raw, test_results_raw = evaluate_model(X_train_raw, y_train_raw, X_test_raw, y_test_raw, dates_test_raw, "No Preprocessing (External)")
```
